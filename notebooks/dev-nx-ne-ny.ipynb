```python
import numpy as np
import matplotlib.pyplot as plt
from reservoir_simulator import ReservoirSim

%load_ext autoreload
%autoreload 2   
    
# Random number generator
rng = np.random.default_rng()
```

```python
nx = [50, 25, 100]
ny = [50, 100, 25]
perm_field = rng.normal(size=nx[0]*ny[0]) + 100

fig, axs = plt.subplots(2, 3, figsize=(12, 8), sharex=True, constrained_layout=True)
for i, (x, y) in enumerate(zip(nx, ny)):
    ph = ReservoirSim(perm_field.reshape((x, y)))()
    axs[0, i].imshow(ph[-1, :, :])
    axs[0, i].set_title(f"nx: {x}; ny: {y}")
    axs[0, i].axis('equal')
    ph[(ph > 149) * (ph < 151)] = np.nan
    axs[1, i].imshow(ph[-1, :, :])
    axs[1, i].set_title(f"nx: {x}; ny: {y}")
    axs[1, i].axis('equal')
#fig.savefig('~/Desktop/comp.png', dpi=300)
```
